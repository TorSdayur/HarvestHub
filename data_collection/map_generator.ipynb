# Folium Demo

dataset was pulled from open nyc, basic idea would be doing this with shelters, food pantrys and any other service that has a physical location or deals with a physical good that must be picked up in person.

```python
import pandas as pd

df = pd.read_csv('Directory_Of_Homeless_Drop-_In_Centers_20240820.csv')
df
```

```python
import folium


map_center = [40.7128, -74.0060]  # Latitude and Longitude of New York City
map = folium.Map(location=map_center, zoom_start=12)

# Iterate through the DataFrame and add markers to the map
for index, row in df.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],  # Adjust column names as needed
        popup=f"{row['Center Name']}<br>{row['Address']}",  # Adjust column names as needed
        icon=folium.Icon(color='blue')
    ).add_to(map)

# Save the map to an HTML file
map.save('homeless_centers_map.html')
```
